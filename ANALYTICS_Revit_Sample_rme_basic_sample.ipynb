```python
# Importing Revit and IFC data
import pandas as pd
df = pd.read_csv('https://DataDrivenConstruction.io/wp-content/uploads/2023/04/rme_basic_sample_project_revit.csv')
df.head()
```

```python
# Filtering data in Revit and IFC projects

df[df['Category'].isin(['OST_Walls', 'OST_Columns'])]
```

```python
# Create a basic pie chart
df['Volume'] = df['Volume'].replace(regex={r' m³': ''}).astype(float)
df.groupby(['Type']).sum().plot.pie(y='Volume', figsize=(17, 17))
```

```python
#Regular expression in Revit and IFC

df[df['Category'].str.match('OST_Wa*')]
```

```python
# Grouping a Revit or IFC project by parameters
df.groupby('Type')['Volume'].sum()
```

```python
#QTO - Finding volumetric quantities for the group

df['Type']= df['Type'].astype(str)
dfq= df[df['Type'].str.match('STB*')]
dfq = dfq.groupby('Type')['Volume', 'Length'].sum()
dfq
```

```python
dfq= df[df['Type'].str.match('STB*')]
```

```python
# Bar plot can be created as follows

dfp = df.groupby('Type')['Volume'].sum()
dfp.plot(kind='barh', figsize=(20, 12))
```

```python
# Grouping with a quantity

df_groups = df.groupby(['Type'])['Volume'].agg(['sum', 'count'])
df_groups
```

```python
# Creating a grouping and saving as Excel

dfe = df.groupby(['Category'])['Length'].agg(['sum', 'count'])
dfe.to_excel("output.xlsx",sheet_name='Category_estimate')  
```

```python
# Replacing values in parameters

df['Type'] = df['Type'].replace("STB d=30", "STB D 30mm")
df
```

```python
# Data filtering option

f_filter = df['Type']=='STB D 30mm'
df[f_filter]
```

```python
from fpdf import FPDF

# Determining the volumetric characteristics of the group
s_cat = 'OST_Walls'
df['Area'] = df['Area'].replace(regex={r' m²': ''}).astype(float)

dfq= df[df['Category'].str.match(s_cat)]
dfq = dfq.groupby('Category')['Volume','Area'].sum()
cat_ar = str(dfq.iloc[0]['Area'])
cat_vol = str(dfq.iloc[0]['Volume'])

# Creating a PDF document based on the parameters found
pdf = FPDF()
pdf.add_page()
pdf.set_font('Arial', 'B', 16)
pdf.cell(190, 8, 'Category: ' + s_cat, 2, 1, 'L')
pdf.set_font('Arial', '', 14)
pdf.cell(190, 8, 'Sum of volumes: ' + cat_vol, 2, 1, 'L')
pdf.cell(190, 8, 'Sum of lengths: ' + cat_ar, 2, 1, 'L')

# Saving a document in PDF format
pdf.output('Report_DataDrivenConstruction.pdf', 'F')
```
